```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
from sklearn.tree import DecisionTreeRegressor  #two lines are procedure
regressor=DecisionTreeRegressor(criterion='poisson', splitter='random', max_features='log2')
regressor=regressor.fit(X_train,y_train)   #fit calculates and creates model and assigns to regressor 
```

```python
import matplotlib.pyplot as plt  #graph purpose
from sklearn import tree
tree.plot_tree(regressor)  #in plot_tree function , regressor has the model so pass it  
plt.show() #shows decision tree graph
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```
